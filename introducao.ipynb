```python
# Initialize a SparkSession

spark = SparkSession.builder.appName("MySparkApp").getOrCreate()
```

```python
bmw_df = spark.read.csv(
    "/home/leonel-cassule/dataEngineer/spark-env/datasets/bmw.csv",
    header=True,
    inferSchema=True
)
```

```python
# Group by company size and calculate the average of salaries
bmw_df.groupBy("Year").agg({"Avg_Price_EUR":"avg"}).show()
```
