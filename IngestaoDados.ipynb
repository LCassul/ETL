```python
# import SparkSession
from  pyspark.sql import SparkSession

# Inition SparkSession

spark =  SparkSession.builder.appName("IngestaoDados").getOrCreate()
```

```python
df_iot = spark.read.csv(
    "datasets/iot-devices.csv",
    header =  True,
    inferSchema = True,
    sep = ";"
)
```

```python
# print schema
df_iot.printSchema()

# show of rows
print(f"Number of rows: {df_iot.count()}")

# show sample records
df_iot.show(5, truncate=False)
```

```python
# Store the first row of df_iot in a variable
first_row = df_iot.first()

# Check the type of the object obtained
print(type(first_row))
```

```python
more_rows = df_iot.take(5)
print(type(more_rows))
```

```python
# import panda, matplotlib
from matplotlib import pyplot as plt


# Convert to Pandas DataFrame

df_pandas =  df_iot.toPandas()

# Create a horizontal bar plot

df_pandas.plot(kind='barh', x = 'battery_level', y = 'c02_level', colormap = 'winter_r')
plt.show()
```
