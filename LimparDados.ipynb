```python
# import sparksession

from pyspark.sql import SparkSession

#Inition  SparkSession

spark = SparkSession.builder.appName("LimparDados").getOrCreate()
```

```python
# Read Dataframe IOT
df_iot = spark.read.csv(
    "datasets/iot-devices.csv",
    header =  True,
    inferSchema = True,
    sep =  ";"
    
)

# Read Dataframe IOT
df_calls = spark.read.csv(
    "datasets/sf-fire-calls.csv",
    header =  True,
    inferSchema = True,
    
)
```
